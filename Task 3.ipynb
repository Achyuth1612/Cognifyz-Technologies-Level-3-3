```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
import sklearn as sk
import seaborn as sns
```

```python
df = pd.read_csv("D:\Cognifyz Internship\Dataset .csv")
df
```

```python
lab=LabelEncoder()
```

```python
df['Has Table booking']=lab.fit_transform(df['Has Table booking'])
df['Has Online delivery']=lab.fit_transform(df['Has Online delivery'])
df
```

```python
total_restaurants = len(df)
table_booking_percentage = (df['Has Table booking'].sum() / total_restaurants) * 100
online_delivery_percentage = (df['Has Online delivery'].sum() / total_restaurants) * 100
total_restaurants
```

```python
print(f"Percentage of restaurants with table booking: {table_booking_percentage:.2f}%")
print(f"Percentage of restaurants with online delivery: {online_delivery_percentage:.2f}%")
```

```python
price_range_avg_rating = df.groupby('Price range')['Aggregate rating'].mean()
highest_avg_rating_color = price_range_avg_rating.idxmax()
highest_avg_rating_color
```

```python
print("\nPrice Range vs. Average Rating:")
print(price_range_avg_rating)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Price range', hue='Has Online delivery', data=df, palette='Set2')
plt.title('Relationship between Price Range and Online Delivery Availability')
plt.xlabel('Price Range')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Price range', hue='Has Table booking', data=df, palette='Set1')
plt.title('Relationship between Price Range and Has Table booking Availability')
plt.xlabel('Price range')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(price_range_avg_rating.index, price_range_avg_rating.values, color='skyblue')
plt.xlabel("Price Range")
plt.ylabel("Average Rating")
plt.title("Average Rating by Price Range")
plt.axhline(price_range_avg_rating[highest_avg_rating_color], color='red', linestyle='--', label="Highest Avg. Rating")
plt.legend()
plt.show()
```
